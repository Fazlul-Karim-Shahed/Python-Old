```python
import pandas
import numpy
from matplotlib import pyplot as plt
from tensorflow import keras
# from sklearn.model_selection import train_test_split
```

```python
ds = pandas.read_csv('../DataSheets/Social_Network_Ads.csv')
ds.head(10)
```

```python
ds = ds.drop(['User ID', 'Gender'], axis=1)
ds
```

```python
plt.scatter(ds.EstimatedSalary, ds.Purchased)
plt.show()
```

```python
ds['Age'] = ds["Age"]/ds.Age.max()
ds['EstimatedSalary'] = ds["EstimatedSalary"]/ds.EstimatedSalary.max()
```

```python
# x_train, x_test, y_train, y_test = train_test_split(ds[['Age', 'EstimatedSalary']], test_size=.2, random_state=25)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(ds[['Age', 'EstimatedSalary']], ds['Purchased'], epochs=1000)
```

```python
model.evaluate(ds[['Age', 'EstimatedSalary']], ds['Purchased'])
```
