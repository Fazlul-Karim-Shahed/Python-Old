```python
import pandas
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import sklearn
```

```python
ds = pandas.read_csv('Datasheets/Maternal Health Risk Data Set.csv')
```

```python
ds.head()
```

# One Hot Encoding

```python
dummy = pandas.get_dummies(ds.RiskLevel)
```

```python
ds = ds.drop('RiskLevel', axis=1)
```

```python
ds = pandas.concat([ds, dummy], axis=1)
```

```python
ds
```

# Ordinal Encoding

```python
ds = pandas.read_csv('Datasheets/Maternal Health Risk Data Set.csv')
```

```python
ds.head()
```

```python
ds.RiskLevel.unique()
```

```python
order = ['low risk', 'mid risk', 'high risk']
```

```python
enc = OrdinalEncoder(categories=[order]).fit_transform(ds[['RiskLevel']])
```

```python
ds.RiskLevel = enc
```

```python
ds
```

# Label Encoding

```python
ds = pandas.read_csv('Datasheets/Maternal Health Risk Data Set.csv')
```
