```python
mean = ds.DiastolicBP.mean()
std = ds.DiastolicBP.std()
pos_ol = mean + (2.5*std)
neg_ol = mean - (2.5*std)
ds[(ds.DiastolicBP > neg_ol) & (ds.DiastolicBP < pos_ol)]
```

```python
mean = ds.SystolicBP.mean()
std = ds.SystolicBP.std()
pos_ol = mean + (2.5*std)
neg_ol = mean - (2.5*std)
ds[(ds.SystolicBP > neg_ol) & (ds.SystolicBP < pos_ol)]
```
